```python
import pandas as pd
```

```python
fastq = pd.read_csv("/Users/martinszyska/mount/snakes/RNAseq/config/fastq.txt", names=['path'], skiprows=1)
```

```python
fastq.head()
```

```python
fastq['sample'] = fastq['path'].str.extract(r"^([A-F][1-6])")
fastq['index'] = fastq['path'].str.extract(r"L4_([12])")
fastq
```

```python
units = fastq.pivot(index="sample", columns="index", values="path").reset_index().rename({
    '1':"fq1", 
    '2':"fq2",
    'sample':"sample_name"
}, axis=1)
units['unit_name'] = 1
units
```

```python
units.iloc[:,[0,-1,1,2]].to_csv("/Users/martinszyska/mount/snakes/RNAseq/config/rnaPol_units.txt", sep="\t", index=False)
```

```python
units
```

```python
samples = units.groupby("sample_name").first().reset_index().iloc[:,[0]]
samples.head(3)
```

```python
samples.loc[:, "condition"] = samples['sample_name'].str.extract(r"[A-Z]([1-9])")
```

```python
samples.to_csv("/Users/martinszyska/mount/snakes/RNAseq/config/rnaPol_samples.txt", sep="\t", index=False)
```
